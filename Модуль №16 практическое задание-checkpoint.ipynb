# Визуализация с seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

## Урок 1. Что такое seaborn

### Практика

#### Задание 1

С помощью seaborn постройте 2 гистограммы распределения минимальных зарплат в 2000 году. Первая гистограмма должна показывать распределение зарплат в прибрежных штатах, а вторая - во внутренних. Гистограммы нужно построить в одних координатных осях.

```python
df = pd.read_csv("wage-data-coast.csv")
df.head()
```

```python
data = df[df['Year'] == 2000]
# data
fig, ax = plt.subplots(figsize=(12, 6))
sns.histplot(data=data, x='Salary', hue='IsCoastal', ax=ax, alpha=0.25, bins=25, multiple='dodge')

ax.set_title('Распределение минимальных зарплат в 2000 году в прибрежных и внутренних штатах')
ax.set_xlabel('Минимальная зарплата, $/час')
ax.set_ylabel('# записей')

plt.show()
```

## Урок 2. Построение основных графиков в seaborn

### Практика

#### Задание 1

Постройте точечный график, на котором по оси х будет указан год, а на оси у - население. Задайте цвет каждой отметки в зависимости от величины минимальной зарплаты в конкретном штате и конкретном году.

```python
df1 = pd.read_csv("wage-data-coast-with-population.csv")
df1.head()
```

```python
data = df1

fig, ax = plt.subplots(figsize=(12,6))
sns.scatterplot(data=data, x='Year', y='Population', ax=ax, hue='Salary', size='Salary')

ax.set_title('Динамика населения США')
ax.set_xlabel('Год')
ax.set_ylabel('Население, млн человек')

plt.show()
```

Постройте FacetGrid, в котором строкам сетки будут соответствовать значения чистоты бриллианта, а столбцам сетки - значения огранки. Внутри сетки постройте точечные графики зависимости цены от высоты бриллианта (это параметр z). Разным цветом на графиках укажите бриллианты разного цвета.
