### Задача 1

Постройте гистограмму распределения объема продаж всех игр в Японии

```python
#df_jp = df[df['JP_Sales'] > 0]['Genre']
#df_jp
#df['JP_Sales']

bin_width = df['JP_Sales'].max() / 10
x_range = [i * bin_width for i in range(11)]

#plt.hist(df_jp)
plt.hist(df['JP_Sales'], log=True)
plt.title('Распределение объема продаж всех игр в Японии')
plt.xlabel('Количество копий, млн')
plt.ylabel('Количество игр, шт.')
plt.xticks(x_range, rotation=45);
```

### Задача 2

Постройте столбчатую диаграмму (barchart), показывающую сравнение общемирового объема продаж игры "Grand Theft Auto V" на разных платформах

```python
df_gtav = df[df['Name'] == 'Grand Theft Auto V'].sort_values('Platform', ascending=False).head(3)
#df_gtav

plt.bar(df_gtav['Platform'], df_gtav['Global_Sales'])
plt.title('Сравнение общемирового объема продаж игры "Grand Theft Auto V" на разных платформах')
plt.xlabel('Платформа')
plt.ylabel('Объём продаж, млн копий')
```

### Задача 3

Постройте линейный график динамики суммарных общемировых продаж всех игр по годам

```python
sum_year = df.groupby('Year').agg({"Global_Sales":["sum"]})
sum_year.index = sum_year.index.astype(int)
#sum_year

plt.plot(sum_year)
plt.title('График динамики суммарных общемировых продаж всех игр по годам')
plt.xlabel('Год')
plt.ylabel('Количество проданных копий, млн')
```

### Задача 4

Постройте точечную диаграмму, показывающую существование зависимости продаж во всем мире от продаж в Северной Америке

```python
df0 = df[df['NA_Sales'] >= 0]
plt.scatter(df0['Year'], df0['NA_Sales'], label = 'продажи в Северной Америке')

df1 = df[df['Global_Sales'] >= 0]
plt.scatter(df1['Year'], df1['Global_Sales'], label = 'продажи в мире')

plt.title('График зависимости продаж во всем мире от продаж в Северной Америке')
plt.xlabel('Год')
plt.ylabel('Количество проданных копий, млн')

plt.legend()
```

### Задача 5

Постройте круговую диаграмму процентного соотношения продаж игры "Super Mario Bros." на разных платформах во всем мире. Подпишите доли графика с точностью 1 знак после запятой.

```python
df_mario_sort = df[df['Name'] == 'Super Mario Bros.'].sort_values('Global_Sales', ascending=False)
df_mario = df_mario_sort[['Platform', 'Global_Sales']].groupby('Platform')['Global_Sales'].sum()

#df_mario

plt.pie(df_mario, labels = df_mario.index, autopct = '%1.1f%%')
plt.title('Соотношение продаж игры "Super Mario Bros." на разных платформах во всем мире')
```

### Задача 6

Определите 5 издателей, имеющих наибольшие суммарные продажи во всем мире в 2013 году и проиллюстрируйте величину их продаж на столбчатой гистограмме

```python
df_2013 = df[df['Year'] == 2013]
df_top = df_2013[['Publisher', 'Global_Sales']].groupby('Publisher')['Global_Sales'].sum()
df_top5 = pd.DataFrame(df_top).sort_values('Global_Sales', ascending=False).head(5)
#df_top5


plt.bar(df_top5.index, df_top5['Global_Sales'], width = 0.5)


plt.title('Сравнение общемирового объема продаж компаний - лидеров по продажам')
plt.xlabel('Компания')
plt.ylabel('Объём продаж, млн копий')
plt.xticks(rotation=45);
```

### Задача 7

Постройте гистограмму распределения величины общемировых продаж игр, выпущенных не издателем Nintendo, в период с 2000 по 2015 год включительно. Гистограмма для большей точности должна содержать 20 столбцов, отсечки на оси X должны соответствовать границам столбцов.

```python
df_2000_2015 = df[(df['Year'] >= 2000) & (df['Year'] <= 2015) & (df['Publisher'] != 'Nintendo')]['Global_Sales']
#df_2000_2015.max()

bin_width = df_2000_2015.max() / 21
x_range = [i * bin_width for i in range(22)]

plt.hist(df_2000_2015, log=True, bins=20)
plt.title('Распределения величины общемировых продаж игр, \n кроме издателя Nintendo, в период с 2000 по 2015 год включительно')
plt.xlabel('Количество копий, млн')
plt.ylabel('Количество игр')
plt.xticks(x_range, rotation=90);
```

### Задача 8

Постройте линейный график динамики суммарных мировых продаж игр жанра "Action" по годам

```python
df_action = df[df['Genre'] == "Action"]
#df_action
sum_year_action = df_action.groupby('Year').agg({"Global_Sales":["sum"]})
sum_year_action.index = sum_year_action.index.astype(int)
#sum_year_action

x_range = list(range(sum_year_action.index.min(), sum_year_action.index.max() + 1, 5))
x_range.append(sum_year_action.index.max())

plt.plot(sum_year_action)
plt.title('График динамики суммарных мировых продаж игр жанра "Action" по годам')
plt.xlabel('Год')
plt.ylabel('Количество проданных копий, млн')
plt.xticks(x_range, rotation=45);
```

### Задача 9

С помощью столбчатой диаграммы проиллюстрируйте объем продаж всех игр издателя "Microsoft Game Studios" в Северной Америке за все время в зависимости от жанра. Столбцы расположите по убыванию.

```python
df_mgs = df[df['Publisher'] == "Microsoft Game Studios"]
#df_mgs
df_mgs_sum = df_mgs[['Genre', 'NA_Sales']].groupby('Genre')['NA_Sales'].sum()
#df_mgs_sum
df_mgs_sum_top = pd.DataFrame(df_mgs_sum).sort_values('NA_Sales', ascending=False)
#df_mgs_sum_top

plt.bar(df_mgs_sum_top.index, df_mgs_sum_top['NA_Sales'])

plt.title('Объем продаж всех игр издателя "Microsoft Game Studios" \n в Северной Америке за все время в зависимости от жанра')
plt.xlabel('Жанр')
plt.ylabel('Объём продаж, млн копий')
plt.xticks(rotation=45);
```

В датасете "vgsales" одним из потенциальных источников ошибок является столбец с общемировыми продажами. По идее, значение общемировых продаж какой-либо игры должно равняться сумме продаж в Северной Америке, Европе, Японии и остальных странах, в противном случае мы получим ошибку в данных, которая может негативно повлиять на весь дальнейший процесс анализа.

1. Ошибка округления. В целом, такая ошибка достаточно часто встречается при переводе величин из одних единиц в другие (например, из тысяч копий в миллионы копий).
