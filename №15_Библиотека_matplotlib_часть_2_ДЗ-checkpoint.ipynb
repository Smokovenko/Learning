```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vgsales.csv')
df.head()
```

### Задача 1

Постройте линейный график динамики объема продаж всех игр жанра "Sports" в Японии. Задайте графику размер 12 на 6

```python
fig, ax = plt.subplots(figsize = (12, 6))

df_sports = df[df['Genre'] == "Sports"][['Year', 'JP_Sales']]
df_sports_japan = df.groupby('Year').agg({'JP_Sales':["sum"]})
# df_sports
# df_sports_japan

ax.plot(df_sports_japan)
ax.set_title('График динамики объема продаж всех игр жанра "Sports" в Японии')
ax.set_xlabel('Год')
ax.set_ylabel('Объём продаж, млн копий')
plt.show()
```

### Задача 2

В одних координатных осях постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
df1 = df[df['Publisher'] == "Activision"][['Year', 'JP_Sales']]
df1_JP = df1.groupby('Year').agg({'JP_Sales':["sum"]})
df2 = df[df['Publisher'] == "Activision"][['Year', 'NA_Sales']]
df2_NA = df2.groupby('Year').agg({'NA_Sales':["sum"]})
df3 = df[df['Publisher'] == "Activision"][['Year', 'EU_Sales']]
df3_EU = df3.groupby('Year').agg({'EU_Sales':["sum"]})
df4 = df[df['Publisher'] == "Activision"][['Year', 'Global_Sales']]
df4_GS = df4.groupby('Year').agg({'Global_Sales':["sum"]})
df2_NA.index = df2_NA.index.astype(int)
# df4
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df1_JP, label='Япония')
ax.plot(df2_NA, label='Северная Америка')
ax.plot(df3_EU, label='Европа')
ax.plot(df4_GS, label='Мир')

ax.set_title('Графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире')
ax.set_xlabel('Год')
ax.set_ylabel('Объём продаж, млн копий')
ax.set_xticks(list(range(df2_NA.index.min(), df2_NA.index.max(), 2)) + [df2_NA.index.max()])
ax.grid()
plt.legend()
plt.show()
```

### Задача 3

В цикле (т.е. на разных координатных осях) постройте линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире

```python
fig, axs = plt.subplots(nrows=4, ncols=1, figsize=(15, 15))

for i, sales in enumerate(['JP_Sales', 'NA_Sales', 'EU_Sales', 'Global_Sales']):
    data = df[df['Publisher'] == "Activision"][['Year', sales]]
    data1 = data.groupby('Year').agg({sales:["sum"]})
    ax = axs[i]
    data1.index = data1.index.astype(int)
    ax.plot(data1)
    ax.set_ylabel('Объём продаж, млн копий; ' + sales)
    ax.grid()
    ax.set_xticks(list(range(data1.index.min(), data1.index.max(), 2)) + [data1.index.max()])
    
    if sales == 'Global_Sales':
        ax.set_xlabel('Год')
        
fig.suptitle('Графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире', y=0.9)
plt.show()
```

### Задача 4

Постройте такие же линейные графики динамики продаж студии "Activision" в Северной Америке, Европе, Японии и всем мире на одном полотне в сетке 2 на 2 графика. Итоговое полотно сделайте размером 12 на 12. Все графики должны разделять оси X и Y

```python
sales = ['NA_Sales','EU_Sales','JP_Sales','Global_Sales']

data = df[df['Publisher'] == 'Activision'].groupby('Year')[sales].sum()

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12,12), sharex=True, sharey=True)

for i, ax in enumerate(axs.ravel()):
    ax.plot(data.index, data[sales[i]], label=sales[i])
    ax.set_title(sales[i] + ': динамика продаж студии "Activision"')
    ax.set_xlabel('Год')
    ax.set_ylabel('Продажи, млн копий')

plt.show()
```

### Задача 5

В одной координатной сетке постройте 2 гистограммы распределения мировых продаж игр издателей "Microsoft Game Studios" и "Take-Two Interactive" после 2010 года (включая 2010 год). Графики сделайте полупрозрачными, отсечки должны быть общими и соответствовать ширине столбцов обоих графиков

```python
df_2010_tti = df[(df['Year'] >= 2010) & (df['Publisher'] == 'Take-Two Interactive')]['Global_Sales']
df_2010_mgs = df[(df['Year'] >= 2010) & (df['Publisher'] == 'Microsoft Game Studios')]['Global_Sales']

fig, ax = plt.subplots(figsize=(12, 6))

_, bins, _ = ax.hist(df_2010_mgs, label = 'Microsoft Game Studios', alpha=0.5, bins=20)
ax.hist(df_2010_tti, label = 'Take-Two Interactive', alpha=0.5, bins=bins)


ax.set_title('Распределение мировых продаж игр издателей "Microsoft Game Studios" и "Take-Two Interactive" после 2010 года')
ax.set_xlabel('Объём продаж, млн копий')
ax.set_ylabel('# записей')
ax.set_xticks(bins)
ax.legend()

plt.show()
```

### Задача 6

Определите 5 платформ, для которых в Японии было продано больше всего игр за все время. Проиллюстрируйте величину продаж на столбчатой диаграмме, столбец с самым высоким значением окрасьте зеленым, а с самым низким - красным

```python
df_platform_jp = df[['Platform', 'JP_Sales']].groupby('Platform')['JP_Sales'].sum()
df_platform_jp_top = pd.DataFrame(df_platform_jp).sort_values('JP_Sales', ascending=False).head(5)
#df_platform_jp_top

fig, ax = plt.subplots(figsize=(12,6))
barlist = ax.bar(df_platform_jp_top.index, df_platform_jp_top['JP_Sales'])
barlist[0].set_color('g')
barlist[4].set_color('r')


ax.set_title('Платформы с максимальным количеством игр в Японии за всё время')
ax.set_xlabel('Платформа')
ax.set_ylabel('Объём продаж, млн копий')


plt.show()
```

### Задача 7

Постройте линейный график динамики количества игр, выпускаемых Nintendo, по годам. С помощью numpy или pandas определите промежуток, в который издатель каждый год выпускал больше 35 игр, и подсветите этот промежуток времени с помощью зеленого полупрозрачного прямоугольника

```python
df_nin = df[df['Publisher'] == 'Nintendo'][['Year', 'Name']].groupby('Year').count()
df_nin_plot = df_nin[df_nin['Name'] >= 35]
df_nin.index = df_nin.index.astype(int)
#df_nin_plot

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(df_nin, linewidth=2)
ax.axvspan(xmin=df_nin_plot.index.min(), xmax=df_nin_plot.index.max(), color='green', alpha=0.5)

ax.set_title('Линейный график динамики количества игр, выпускаемых Nintendo, по годам')
ax.set_xlabel('Год')
ax.set_ylabel('Количество выпущенных игр, шт.')
ax.set_xticks(list(range(df_nin.index.min(), df_nin.index.max(), 3)) + [df_nin.index.max()])
ax.grid()
plt.show()
```

### Задача 8

Определите 3 жанра и 4 издателя с самыми большими продажами игр в мире за все время. Постройте сетку графиков 4 на 3, в каждой ячейке постройте точечный график, проверяющий зависимость общемировых продаж от продаж в Северной Америке для каждой пары (жанр, издатель). Размер итогового полотна определите самостоятельно таким образом, чтобы все графики были видны

```python
top_genres = df.groupby('Genre')['Global_Sales'].sum().nlargest(3).index
top_publishers = df.groupby('Publisher')['Global_Sales'].sum().nlargest(4).index

fig, axs = plt.subplots(nrows=len(top_publishers), ncols=len(top_genres), figsize=(12, 12))
for i, pub in enumerate(top_publishers):
    for j, genre in enumerate(top_genres):
        ax = axs[i][j]
        data = df.query('Genre == @genre and Publisher == @pub')[['Global_Sales', 'NA_Sales']]
            
        ax.scatter(data['NA_Sales'], data['Global_Sales'])
        
        if i == len(top_publishers) - 1:
            ax.set_xlabel('Продажи в Северной Америке, млн копий')
        if j == 0:
            ax.set_ylabel('Мировые продажи, млн копий')
plt.suptitle('График, проверяющий зависимость общемировых продаж от продаж в Северной Америке для пары жанр/издатель', y=0.93)
plt.show()
```
