### Вопросы по лекциям.

Как поймать вообще все ошибки, которые могут произойти?

**Ответ:**
Код который может содержать ошибки поместить в каплусу: try/except.

Сколько раз подряд можно указывать except?

**Ответ:**
Можно перечислять классы ошибок без неограниченное число раз, но только если их обработка одинаковая.

Вы хотите с помощью print вывести название ошибки в консоль, как это сделать?

**Ответ:**
Указать ошибку как {exc}. Например: 
except ZeroDivision as exc:
    print(".....{exc}....")

Вы хотите с помощью print вывести параметры ошибки в консоль, как это сделать?

**Ответ:**
Указать ошибку как {exc} и указать параметры ошибки как {exc.args}. Например: 
except OSError as exc:
    print(".....{exc}....параметры {exc.args}")

Что такое DeprecationWarning?

**Ответ:**
Уведомление о том, что используемая часть кода (функционала) морально устарела/была заменена.

 - присутсвуют все три поля
 - поле имени содержит только буквы
 - поле email содержит @ и .
 - поле возраст является числом от 10 до 99<br>

 - registrations_good.log для правильных данных, записывать строки как есть

Для валидации строки данных написать метод, который может выкидывать исключения:
 - НЕ присутсвуют все три поля: ValueError
 - поле имени содержит НЕ только буквы: NotNameError (кастомное исключение)
 - поле email НЕ содержит @ и .(точку): NotEmailError (кастомное исключение)
 - поле возраст НЕ является числом от 10 до 99: ValueError
Вызов метода обернуть в try-except.

```python
### YOUR CODE HERE ###

class NotNameError(Exception):
    pass

class NotEmailError(Exception):
    pass

class AgeError(Exception):
    pass

def func_line(line):
    name, email, age = line.split(" ")
    name_user = name
    email_user = email
    age_user = age
    if str(name_user).isalpha() is not True:
        raise NotNameError("NotNameError")     
    elif ("@" not in str(email_user)) or ("." not in str(email_user)):
        raise NotEmailError("NotEmailError")
    elif int(age_user) < 1 or int(age_user) > 99:
        raise AgeError("AgeError")
    else:
        return line

import os
count_all = 0
count_good = 0
count_bad = 0 
count_ValueError = 0
count_NotNameError = 0
count_NotEmailError = 0
count_AgeError = 0
with open(os.path.join("d:\Папка моя\Пежня\Скоро будет удалено\Новая папка\Модуль №9", "registrations_.txt"), "r", encoding="utf-8") as ff, open(os.path.join("d:\Папка моя\Пежня\Скоро будет удалено\Новая папка\Модуль №9", "registrations_bad.log.txt"), "w", encoding="utf-8") as bb, open(os.path.join("d:\Папка моя\Пежня\Скоро будет удалено\Новая папка\Модуль №9", "registrations_good.log.txt"), "w", encoding="utf-8") as gg:
    for line in ff:
        line = line[:-1]
        line = str(line + "\n")
        count_all += 1
        try:
            func_line(line)
            gg.write(line)
            count_good += 1
        except ValueError as exc:
            if "unpack" in exc.args[0]:
                print(f"Не хватает операндов! Ошибка: {exc}, в строке {line}")
                bb.write(f"Не хватает операндов! Ошибка: {exc}, в строке {line}")
                count_bad += 1
                
        except NotNameError as exc:
            print(f"Неверно указано имя! Ошибка: {exc}, в строке {line}")
            bb.write(f"Неверно указано имя! Ошибка: {exc}, в строке {line}")
            count_bad += 1
            
        except NotEmailError as exc:
            print(f"Неверно указан email! Ошибка: {exc}, в строке {line}")
            bb.write(f"Неверно указан email! Ошибка: {exc}, в строке {line}")
            count_bad += 1
        
        except AgeError as exc:
            print(f"Неверно указан возраст! Ошибка: {exc}, в строке {line}")
            bb.write(f"Неверно указан возраст! Ошибка: {exc}, в строке {line}")
            count_bad += 1
            
        finally:
            continue
print(count_all)
print(count_good)
print(count_bad)


```
